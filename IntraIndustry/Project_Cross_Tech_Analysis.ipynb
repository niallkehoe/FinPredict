```python
import pandas as pd

techCompanies = ["AAPL", "AMZN", "GOOG", "META", "MSFT", "NVDA", "PYPL", "TSLA"]

df = pd.read_csv("/"+techCompanies[0]+".csv", index_col=0)
# df = pd.DataFrame()
```

```python
for i in range(1,len(techCompanies)):
  dfNew = pd.read_csv("/"+techCompanies[i]+".csv", index_col=0)
  df = df.merge(dfNew, on="Date", suffixes=["_"+techCompanies[i-1],"_"+techCompanies[i]])
```

```python
df
```

```python
columns = ["Close_"+a for a in techCompanies]
columns.append("Date")
dfPrices = df[columns]
dfPrices
```

```python
dfPrices.describe()
```

```python
from matplotlib import pyplot
# Analysis from https://github.com/tatsath/fin-ml/blob/master/Chapter%205%20-%20Sup.%20Learning%20-%20Regression%20and%20Time%20Series%20models/Case%20Study%201%20-%20Stock%20Price%20Prediction/StockPricePrediction.ipynb
dfPrices.hist(bins=50, sharex=False, sharey=False, xlabelsize=1, ylabelsize=1, figsize=(12,12))
pyplot.show()
```

```python
# Analysis from https://github.com/tatsath/fin-ml/blob/master/Chapter%205%20-%20Sup.%20Learning%20-%20Regression%20and%20Time%20Series%20models/Case%20Study%201%20-%20Stock%20Price%20Prediction/StockPricePrediction.ipynb
dfPrices.plot(kind='density', subplots=True, layout=(4,4), sharex=True, legend=True, fontsize=1, figsize=(15,15))
pyplot.show()
```

```python
# Analysis from https://github.com/tatsath/fin-ml/blob/master/Chapter%205%20-%20Sup.%20Learning%20-%20Regression%20and%20Time%20Series%20models/Case%20Study%201%20-%20Stock%20Price%20Prediction/StockPricePrediction.ipynb
import seaborn as sns
corr = dfPrices.corr()
pyplot.figure(figsize=(18,18))
pyplot.title('Correlation Matrix')
sns.heatmap(corr, vmax=1, square=True, annot=True,cmap='cubehelix')
```
